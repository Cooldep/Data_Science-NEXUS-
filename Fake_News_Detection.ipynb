```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
#Importing the Dataset
path1='/content/drive/MyDrive/Fake News Dataset/Fake.csv'
path2='/content/drive/MyDrive/Fake News Dataset/True.csv'
data_fake = pd.read_csv(path1)
data_true=pd.read_csv(path2)


```

```python
#Assigning Classes to the Dataset

data_fake["class"]=0
data_true["class"]=1
```

```python
#Manual Testing for Both the Dataset

data_fake_manual_testing= data_fake.tail(10)
for i in range(23480,23470,-1):
    data_fake.drop([i], axis = 0, inplace = True)


data_true_manual_testing= data_true.tail(10)
for i in range(21416,21406,-1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
#Merging Both the Dataset

data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
#Dropping Unwanted Columns

data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data= data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
   text = text.lower()
   text = re.sub('\[.*?\]', '', text)
   text = re.sub("\\W", " ", text)
   text = re.sub('https?://\S+|www\.\S+', '', text)
   text = re.sub('<.*?>+', '', text)
   text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
   text = re.sub('\n','', text)
   text = re.sub('\w*\d\w*', '', text)
   return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
#Applying Function to Text Column and Assigning X and Y

x = data['text']
y = data['class']
```

```python
# Defining Training and Testing Data and Splitting Them Into &5 -25 Percent Ratio

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
data.head(10)
```

```python
# Converting Raw Data Into Matrix for Further Process

from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
# Creating First Model

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

**Checking the Model Accuracy and Classification Report**

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

**Creating a Second Model.**

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

**Checking the Model Accuracy and Classification Report**

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n==1:
    return "Not a Fake News"

def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_DT = DT.predict(new_xv_test)
  return print("\n\nLR Predction: {} \nDT Predction:".format(output_lable(pred_LR[0]),output_lable(pred_DT[0])))
```
